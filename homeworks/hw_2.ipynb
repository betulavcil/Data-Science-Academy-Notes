```python
base_url = "C:/Users/avcil/projects/data-science/homeworks/data/covid/"

infected_dataset_url = base_url + "time_series_covid19_confirmed_global.csv"
recovered_dataset_url = base_url + "time_series_covid19_recovered_global.csv"
deaths_dataset_url = base_url + "time_series_covid19_deaths_global.csv"
```

```python

infected_data = pd.read_csv(infected_dataset_url)
recovered_data = pd.read_csv(recovered_dataset_url)
deaths_data = pd.read_csv(deaths_dataset_url)
```

```python

infected_data = infected_data.drop(columns =["Province/State", "Lat", "Long"])
recovered_data = recovered_data.drop(columns =["Province/State", "Lat", "Long"])
deaths_data = deaths_data.drop(columns =["Province/State", "Lat", "Long"])

#drop fonksiyonu ile istemediğimiz columnları düştük  // ben bu columnların gerekli olmadığını düşündüm
# iloc fonksiyonu ile istediğimiz columnları çekebilirdik istemediğimiz daha azdı o yüzden drop ile yapmayı tercih ettik
```
